## <font color='blue'> <b>itemsize</b><font color='black'>
    
itemsize() shows how many bytes each element of the NumPy array occupies in memory.

This property depends on the data type (dtype) of the NumPy array and varies with the size of this data type.

itemsize() can also be used to calculate the total space a NumPy array takes up in memory. This calculation can be done by multiplying the number of elements of the array by itemsize().

The float32 type uses 32 bits (4 bytes) of memory and can represent decimal numbers with approximately 7 digits of precision.

A 32-bit number can represent 2^32 different numbers.

The float64 type uses 64 bits (8 bytes) of memory and can represent decimal numbers with approximately 15-16 digits of precision.

A 64-bit number can represent 2^64 different numbers.

## <font color='blue'> <b>copy() </b><font color='black'>
copy() is used to create a copy in Numpy arrays.

This function creates an exact copy of the original array, but this copy is independent. That is, any changes made to the copy sequence do not affect the original sequence, and changes made to the original sequence do not affect the copy sequence.

# <font color='green'> <b>Concatenation of the Arrays</b><font color='black'>
concatenate() function is used to concatenate two or more arrays.

The first parameter of the function is a list of arrays to be concatenated.

The second parameter is a parameter that specifies the axis or dimension according to which the merging process will be performed.

For example, the axis=0 parameter allows arrays to be concatenated row-wise (vertical join), while the axis=1 parameter allows column-wise concatenation (horizontal join).



```python
# axis = 0 -> rows , 
# if you would like to concatenate the rows, your arrays column numbers should be same 
# if you would like to concatenate the columns , your arrays row numbers should be same 
# otherwise you cant concatenate
```

```python
arr1 = np.arange(1,7) # 1D
arr1
```

```python
arr1.ndim
```

```python
arr2 = np.arange(7,13)
arr2
```

```python
arr2.ndim
```

```python
arr1_2 = np.concatenate((arr1,arr2))
arr1_2# we can only concatenate 1d arrays next to each other 
```

```python
# in one dimensional arrays we cant talk about rows and columns 
# so if it is one dimensional we cant create a row in it 
```

```python
arr12 = np.concatenate((arr1,arr2), axis = 1) # axis 1 -> column
arr12 # ERROR
```

```python
a = np.array([[1,2],[3,4]])
a
```

```python
b = np.array([[5,6],[7,8]])
b
```

```python
a_b = np.concatenate((a,b))
a_b

## the row numbers is increased finally 
# as default axis is 0 -> which means rows
```

```python
a_b = np.concatenate((a,b), axis = 1)
a_b
# finally column number is increased , 
# axis 1 stands for column 
```

```python
# when we use axis 0 it means it puts the arrays bottom of each other -> rows increased
# when we use axis 1 it means it puts the arrays next to each other -> columns increased
```

```python
x = np.array([[1,2],[3,4]])
x
```

```python
y = np.array([[5,6],[7,8],[9,10]])
y
```

```python
xy = np.concatenate((x,y),axis = 0)
xy
```

```python
xy = np.concatenate((x,y),axis = 1)
xy # ERROR  -> when we concatenate in axis 1 level , row number should be same of our arrays
```

```python
xy = np.concatenate((x,y),axis = None)
xy # axis None creates a 1D array including all elements of both arrays as a result 
```

**split():** is used to split the numpy array into equal parts along the specified axis.

The **array_split():** function is used in NumPy to split an array into a specified number of unequal parts. This function, unlike other split functions (such as hsplit, vsplit), divides into parts specified by the user, not on a specific axis.

***vsplit():** Used in NumPy to split an array vertically (i.e. across lines) into multiple parts.

***hsplit():** Used in NumPy to split an array horizontally (i.e. across columns) into multiple parts. Essentially, it splits a matrix at specified indices using its columns.

To access a specific element in NumPy arrays, we specify the index number of the relevant element in square brackets following the array name. For example; myarr[indexno]

Index numbers start from 0, that is, the index of the first element is 0 and the index numbers of the other elements increase respectively by 1, 2, 3, etc.

We can also use negative indexing. Negative indexing is used to access elements in NumPy arrays starting from the end. The index of the last element is -1, and the indexes of the previous elements are -2, -3, etc.

In 2-dimensional NumPy arrays, we use row and column numbers to access a particular element. In this way, myarr[linenumber, columnnumber] format is used for access.

In 3D NumPy arrays, we use depth, row and column numbers to access the element. That is, myarr[depthno, rowno, columnno] format is used for access.

Broadcasting is a feature used to perform arithmetic operations on NumPy arrays of different shapes.

This feature allows NumPy to perform the operation more flexibly by automatically executing the operation in cases where the shapes of two arrays must match exactly to perform the operation.

The broadcasting process is performed between a smaller sized array and a larger sized array.

In this process, the smaller sized array is adapted to the other array and made operable.

```python
import numpy as np
```

```python
np.ones((5,5)) + 6
```

```python
np.zeros((3,3)) 
```

```python
np.arange(0,3)
```

```python
np.zeros((3,3)) + np.arange(0,3)
```

```python
np.zeros((3,3)) + np.arange(0,5) # ERROR -> the first one is 3 x 3 matrix , second one is a vector which has 5 elements
# their shapes are not proper for broadcasting 
```

# <font color='green'> <b>Arithmetic Operations & Universal Array Functions</b><font color='black'>

Arithmetic Operations are functions used to perform arithmetic operations on NumPy arrays.

By vectorizing mathematical operations, it enables fast and effective calculations by avoiding slow operations such as for loops.

It supports many arithmetic operations such as addition, subtraction, multiplication, division, exponentiation, square root and so on.

It can be implemented on NumPy arrays of different sizes and shapes.

Universal Array Functions (Ufuncs) are functions used to perform fast and efficient mathematical operations on NumPy arrays. These functions are specifically designed to perform fast calculations on large data sets.

Ufuncs functions can be used to vectorize mathematical operations. In this way, it is possible to perform fast operations on NumPy arrays without dealing with for loops and similar operations.

NumPy Ufuncs functions include many functions such as trigonometric operations (sin, cos, tan, etc.), logarithmic operations (log, exp, etc.), arithmetic operations (addition, subtraction, multiplication, division, etc.). These functions can be used on NumPy arrays of different sizes and shapes.
